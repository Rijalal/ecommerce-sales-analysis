# Olist E-commerce — Business Insights Report

**Dataset:** 115,000+ delivered orders · Brazil · 2016–2018  
**Tools:** Python (pandas, matplotlib, seaborn)  
**Author:** M. Rijal Alghafiqih

## Executive summary

Analysis of 115,000+ completed Olist orders identifies five actionable 
insights across revenue, logistics, and customer satisfaction. 
Key finding: delivery speed is the single biggest driver of review scores, 
and November represents a predictable revenue opportunity that sellers 
are likely underprepared for.

## Insight 1 — November is the revenue peak, not the average

**Finding:** Revenue in November 2017 spiked significantly above the 
monthly average, consistent with Black Friday and year-end shopping behavior.

**Implication:** Sellers who treat November like a normal month are leaving 
money on the table. Inventory should be increased in October to meet demand.

**Recommendation:** Olist should send sellers a pre-November restocking 
alert in the first week of October, targeting the top 10 revenue categories.

## Insight 2 — Home & bath products dominate revenue

**Finding:** `bed_bath_table` is the single highest revenue-generating 
category across all 115,000+ orders, outperforming electronics and 
other high-ticket categories.

**Implication:** This category deserves priority treatment — more seller 
recruitment, better search placement, and dedicated promotional campaigns.

**Recommendation:** Olist should increase seller acquisition efforts 
specifically in the home & bath segment and analyze whether supply 
is keeping up with demand in this category.

## Insight 3 — Delivery speed directly drives customer satisfaction

**Finding:** Orders with long delivery times (20+ days) consistently 
receive 1–2 star reviews, while fast deliveries receive 4–5 stars. 
The relationship is clear and consistent across all score levels.

**Implication:** Delivery time is not just a logistics metric — it is 
the primary driver of customer satisfaction on the platform.

**Recommendation:** Olist should flag orders that are at risk of exceeding 
15 days by day 10, and proactively notify customers to set expectations 
and prevent negative reviews.

## Insight 4 — Northern states have a logistics problem

**Finding:** Alagoas (AL) has the worst on-time delivery rate on the 
platform at just below 80%, meaning 1 in 5 orders arrives late. 
Northern and northeastern states consistently underperform compared 
to São Paulo and the south.

**Implication:** Customers in these regions have a systematically worse 
experience, which likely suppresses repeat purchase rates.

**Recommendation:** Olist should partner with regional logistics providers 
in the north and northeast, or set adjusted delivery time expectations 
for customers in these states to reduce review score damage.

## Insight 5 — Credit card is the dominant payment method

**Finding:** The vast majority of Olist buyers pay by credit card, 
with installment payments being a key feature — consistent with 
Brazilian consumer behavior where parcelamento (installments) 
is culturally standard.

**Implication:** Any friction in the credit card checkout flow 
directly impacts the majority of transactions on the platform.

**Recommendation:** Olist should prioritize credit card checkout 
optimization and prominently feature installment options at checkout, 
particularly for higher-ticket categories like furniture and electronics.

## Conclusion

These five insights point to three priority areas for Olist:

1. **Logistics** — fix northern state delivery times and flag at-risk orders early
2. **Seasonality** — prepare sellers for November demand spikes proactively  
3. **Category focus** — double down on home & bath as the platform's revenue engine
