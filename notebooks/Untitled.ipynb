```python
import pandas as pd

df = pd.read_csv('../data/clean.csv')

# Check the actual average
print("Average delivery days:", df['delivery_days'].mean())
print("Total rows:", len(df))

# Check if there are NaN values being ignored
print("NaN values in delivery_days:", df['delivery_days'].isna().sum())

# Check a sample of the actual values
print("\nSample of delivery_days values:")
print(df['delivery_days'].head(20).tolist())
```

```python
import pandas as pd

df = pd.read_csv('../data/clean.csv')

# Check if delivery_days column has any huge values hiding
print("Value counts of delivery_days > 50:")
print(df[df['delivery_days'] > 50]['delivery_days'].value_counts().head(20))

# Check total sum and count
print("\nSum:", df['delivery_days'].sum())
print("Count:", df['delivery_days'].count())
print("Manual average:", df['delivery_days'].sum() / df['delivery_days'].count())
```

```python
import pandas as pd

df = pd.read_csv('../data/clean.csv')

print("Average:", df['delivery_days'].mean())
print("Median:", df['delivery_days'].median())
print("\nDistribution:")
print("0-20 days:", len(df[df['delivery_days'] <= 20]))
print("21-50 days:", len(df[(df['delivery_days'] > 20) & (df['delivery_days'] <= 50)]))
print("51-100 days:", len(df[(df['delivery_days'] > 50) & (df['delivery_days'] <= 100)]))
print("100+ days:", len(df[df['delivery_days'] > 100]))

# Check the actual CSV file directly
print("\nFirst 5 delivery_days values in CSV:")
print(df['delivery_days'].head())
```
